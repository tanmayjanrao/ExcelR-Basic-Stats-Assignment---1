# Data Visualization


Objective: To visualize the distribution and relationship of numerical and categorical variables in the dataset.

```python
df.columns
```

```python
numeric_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
df2 = df[numeric_columns]

df2.head()
```

```python


fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 8))

# Plot each histogram with values displayed
for i, col in enumerate(df2.columns):
    row = i // 2  # Calculate row index
    col_idx = i % 2  # Calculate column index
    
    ax = axes[row, col_idx]
    df[col].hist(ax=ax, grid=True)  # Plot histogram
    
    # Add values to the bars
    for p in ax.patches:
        ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
    ax.set_title(col)  # Set title for each subplot

plt.tight_layout()
plt.show()
```

```python
# [1] we can see almost every histogram plot is positively skewed
# [2]check Discount Rate (%) histogram few products have high discount that can be because these products with lower demand 
# or nearing their expiration date or part of promotional campaigns. And the left part of that column Discount Rate (%) 
# are products sold with near to no discount these products are in high demand thats why the discount given is also less

# 


```

```python
brand_counts = df3['Brand'].value_counts()
plt.bar(brand_counts.index, brand_counts.values)

plt.xlabel('Brand')
plt.ylabel('Number of SKUs')
plt.title('Count of SKUs by Brand')
plt.tight_layout()
plt.show()


total_skus = len(df3)
print(f"Total SKUs: {total_skus}")
for brand, count in brand_counts.items():
    percentage = (count / total_skus) * 100
    print(f"{brand}: {count} SKUs ({percentage:.2f}%)")
    
    
# we take category on x axis and numbers on y axis
```

```python
model_counts = df3['Model'].value_counts()

# Create a bar plot for Model distribution
plt.figure(figsize=(10, 6))
plt.bar(model_counts.index, model_counts.values, color='lightgreen')
plt.xlabel('Model')
plt.ylabel('Count')
plt.title('Count of Each Model in City C')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


total_models = len(df3)
print(f"Total Models: {total_models}")
for model, count in model_counts.items():
    percentage = (count / total_models) * 100
    print(f"{model}: {count} occurrences ({percentage:.2f}%)")
```

# Standardization of Numerical Variables

```python
copied_df = df2.copy()
copied_df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaleStandard= StandardScaler()
```

```python
df_standardize = scaleStandard.fit_transform(copied_df)
```

```python
standardize = pd.DataFrame(df_standardize, columns= ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value'])
```

Show before and after comparisons of the data distributions.

```python
standardize.head()  # we got out standardized value here
```

```python
df2.head()   # these were the original values
```

```python
# Standardization is a technique in which for ex- we have a dataset in which the height feauture is measured in inches and
# other feauture which is weight is measured in pounds so now the problem with such data is that the features and not in the 
# same scale so because of this our model can get biased towards values which are greater and the accuracy on test data
# wont be great.. so to avoid such situation we use Standardization which only used for features(columns) and 
# normalization is used for rows 
# in Standardization we use formula z=x-mu/sigma 
# AFTER STANDARDIZATION we get mean approximately close to zero and and standard deviation as 1 you can see below
# sigma is for standard deviation which can be calculated based on population or sample 
```

```python
standardize.describe().round(3)
```

Early we had problem looking at the boxplots now sinse we have standardized out features lets check if there any difference

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=standardize)

plt.xticks(rotation=45)  # Rotate labels by 45 degrees
plt.tight_layout()  # Adjust layout to make room for the rotated labels
plt.show()

# these boxplots are way better and we easily spot outliers and visualise the data distribution
#Central Tendency: The position of the median line within the box indicates the central tendency of the data. 
# If the median is closer to Q1, the data may be left-skewed; if its closer to Q3, it may be right-skewed.
# as observed earlier the Discount Rate(%) feature is right-skewed

#A wider box indicates greater variability in the data, while a narrower box suggests less variability
```

# Conversion of Categorical Data into Dummy Variables

```python
df3.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse_output=False).set_output(transform = 'pandas')
```

```python
ohetransform = ohe.fit_transform(df[['BU']])
ohetransform
```

```python
df4 = pd.concat([df3, ohetransform], axis=1).drop(columns = ['BU'])
```
